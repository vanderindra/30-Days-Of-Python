# Using range() to Make a Lists of Numbers

```python
numbers = list(range(1,6))
print(numbers)
```

```python
even_numbers = list(range(2,11,2))
print(even_numbers)
```

```python
squares = []
for value in range(1,11):
    square = value ** 2
    squares.append(square)
```

```python
print(squares)
```

# More Concisely

```python
squares = []
for value in range(1,11):
    squares.append(value ** 2)
    
print(squares)
```

# List Comprehensions

```python
squares = [value ** 2 for value in range(1,11)]
print(squares)
```

```python
numbers = []

for value in range(1,20):
    numbers.append(value)
    
print(numbers)
```

```python
millions = []

for value in range(1,100):
    millions.append(value)
    print(millions)
```

```python
numbers_millions = []

for value in range(1,1000):
    numbers_millions.append(value)
print(numbers_millions)
```

```python
min(numbers_millions)
```

```python
max(numbers_millions)
```

```python
sum(numbers_millions)
```

```python
odd_number = []

for value in range(1,20,2):
    #if value % 2 != 0:
    odd_number.append(value)
    print(odd_number)
```

```python
power_3 = []

for value in range(3,30):
    multiple = value * 3
    power_3.append(multiple)
    print(power_3)
```

```python
cube = []

for value in range(1,10):
    power_3 = value ** 3
    cube.append(power_3)
    print(cube)
```

```python
cube = [value ** 3 for value in range(1,10)]
print(cube)
```
